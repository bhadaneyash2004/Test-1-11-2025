```python
#Q6
def sum_n(n):
    if n == 0:
        return 0
    else:
        return (n % 10) + sum_n(n//10)
print(sum_n(127))
```

```python
#Q8
def analyze_numbers(*args):
    sums = sum(args)
    average = sums / len(args)
    smallest = min(args)
    largest = max(args)
    return sums,average,smallest,largest
analyze_numbers(1,2,3,4)
```

```python
#Q3
def stats_summary(*args):
    count = 0
    average = sum(args) / len(args)
    smallest = min(args)
    largest = max(args)
    counts = len(args)
    dict = {"Min":smallest,"Max":largest,"Mean":average,"count":counts}
    return dict
stats_summary(1,2,3,22,4,5)
```

```python
#Q9
def splite_even_odd(*number):
    even = []
    odd = []
    for num in number:
        if num % 2 == 0:
            even.append(num)
        else:
            odd.append(num)
    return even,odd
splite_even_odd(2,3,4,5,6)
```

```python
#12
def validate_email(email):
    if '@' in email:
        if ".com" in email or ".in" in email:
            return True
        else:
            return False
    return False
validate_email("xyz@gmail.in")
```
