```python
#1. What is the overall distribution of order values across all categories?
group_sum = df.groupby("Product_Category")["Order_Value"].sum()
plt.bar(group_sum.index ,group_sum.values)
plt.title("overall distribution of order values")
plt.xlabel("Product Category")
plt.ylabel("Order Values")
plt.show()
```

```python
#2. How many orders result in refunds compared to total transactions?
```

```python
#3. Which product categories show the highest frequency of returns?
plt.hist(df["Product_Category"],color = "green")
plt.title("product categories show the highest frequency")
plt.xlabel("Product Category")
plt.ylabel("Frequency")
plt.show()
```

```python
#4. What is the relationship between discount percentage and return rate?
plt.scatter(df['Return_Request_Time_Days'],df['Discount_Applied_pct'],color="red",edgecolor = "black")
```

```python
# 5. Do higher discounts correlate with more frequent refund requests?
corr = df[['Discount_Applied_pct','Return_Request_Time_Days']].corr()
sns.heatmap(corr)
```

```python
#7. Which subcategories have the most refund-related complaints?
plt.bar(df['Sub_Category'],df['Refund_Amount'],color="green")
plt.title("subcategories vs refunds")
plt.xlabel("subcategories")
plt.ylabel("Refund Amount")
plt.show()
```

```python
#8. Are specific age groups more likely to request refunds?
age_group = df.groupby('Customer_Age')['Refund_Amount'].sum()
plt.bar(age_group.index,age_group.values, color = "red")
plt.xlabel("Customer Age")
plt.ylabel("Refund Amount")
plt.show()
```
