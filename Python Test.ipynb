## Q2. Write python a function that checks whether all numbers in a list are greater than 10 using inbuilt functions.

```python
def greater(ls):
    result = list(filter(lambda x: x > 10 ,ls))
    if len(ls) == len(result):
        return True
    else:
        return False
ls = [20,11,11,22,44,55,67]
print(grater(ls))
```

## Q10. For the following dataframe of transactions Plot the total amount spent by each customer using a horizontal bar chart (ordered by value).

```python
import pandas as pd 
import  matplotlib.pyplot as plt
df = pd.DataFrame({
    'CustomerID': [101, 102, 101, 103, 102, 104],
    'Amount': [150, 90, 200, 300, 120, 80],
    'TransactionDate': pd.to_datetime([
        '2023-01-01', '2023-01-02', '2023-01-10', 
        '2023-02-01', '2023-02-10', '2023-03-01'
    ])
})
df.head()
```

```python
group = df.groupby('CustomerID')['Amount'].sum()
print(group)
```

```python
plt.barh(group.index,group.values, color = "blue",edgecolor = 'black')
plt.title("Total Amount Spent by each customer")
plt.xlabel("Amount")
plt.ylabel("Customers")

plt.show()
```

## Q11. For the dataframe of transactions (above mentioned df).
- Calculate total sales by region

- Plot as a pie chart with percentages and exploded slices

```python
df = pd.DataFrame({
    'CustomerID': [101, 102, 101, 103, 102, 104],
    'Amount': [150, 90, 200, 300, 120, 80],
     'Region':['Pune','Nashik','Mubai','Nagpur','Amravati','Aurangabad'],
    'TransactionDate': pd.to_datetime([
        '2023-01-01', '2023-01-02', '2023-01-10', 
        '2023-02-01', '2023-02-10', '2023-03-01'
    ])
})
```

```python
group = df.groupby('Amount')['Region'].sum()
print(group)
```

```python
plt.pie(group.index,labels = group.values,autopct = "%1.1f%%")
plt.show()
```
